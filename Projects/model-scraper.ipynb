```python
# Import all the things!
import urllib.request
from datetime import *
from lxml import html
from bs4 import BeautifulSoup

# Scrape all HTML from webpage.
def scrapewebpage(url):
	# Open URL and get HTML.
	web = urllib.request.urlopen(url)

	# Make sure there wasn't any errors opening the URL.
	if (web.getcode() == 200):
		html = web.read()
		return(html)
	else:
		print("Error %s reading %s" % str(web.getcode()), url)

# Helper function that scrape the webpage and turn it into soup.
def makesoup(url):
	html = scrapewebpage(url)
	return(BeautifulSoup(html, "lxml"))
```

```python
# Scrape page with list of all models.
models_soup = makesoup("http://www.elitemodel.se/women/direct/")

# Get list of all models in two steps.
# First, we get the <div class="search-results"> and then the <ul> inside.
div = models_soup.find("div", "search-results")
models = div.find("ul")
```

```python
# Create an empty list were we'll store the links.
links = []

# Get each link and add it to the end of the list.
for model in models.find_all("a"):
    # Make sure only women models are added.
    if model["href"][0:6] == "/women":
        # Add model to list.
        links.append("http://www.elitemodel.se" + model["href"])
    
```

```python
# We have to import "time" to be able to add delays.
import time
```

# Note:

This will scrape very slowly - and that's the point. Too fast, and the server will get angry and overload.

A **1 second delay** between each scrape is sufficient.

```python
# A counter that stores how many pages (models) scraped in total.
i = 0

# The next step is to go through all links in the list "links".
for link in links:
    # Increase counter everytime we scrape a page.
    i = i + 1
    
    print("Scraping {0} of {1} ({2}%): {3}".format(i, len(links), round((i/len(links)) * 100, 1), link))
    
    # Scrape specific model page.
    data_soup = makesoup(link)

    # Extract data for measurements.
    measurements = data_soup.find("div", "dimensions")
    p = measurements.find_all("p")
    print(p[0].text, p[1].text, p[2].text, p[3].text, p[4].text, p[5].text, p[6].text)

    # Delay one second.
    time.sleep(1.0)
        
    # This will stop the scraping after five pages. This is useful during
    # development so you can see that everything seems to be working.
    # Remove when you want to scrape all pages.
    if i > 4:
        print("Reached max")
        break
        
print("Done")
```
